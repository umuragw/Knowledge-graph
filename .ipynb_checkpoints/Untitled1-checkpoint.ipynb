```python
import csv

# 将标题和内容写入 CSV 文件
csv_file = 'data.csv'  # CSV 文件名

# 打开 CSV 文件，选择写入模式
with open(csv_file, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)

    # 写入标题和内容作为一行数据
    writer.writerow(['标题', '内容'])
    for i in range(len(l1)):
        writer.writerow([l1[i], l2[i]])

print(f'标题和内容已写入到 {csv_file} 文件中。')
```
